```python
import numpy as np
import scipy.signal

def chirp(tmax=0.001, f1=22000, f2=1000, fs=44100):
    f = np.exp(np.linspace(np.log(2 * np.pi * f1/fs), np.log(2 * np.pi * f2/fs), int(tmax * fs)))
    amp = np.tanh(np.sin(np.linspace(0,np.pi,len(f)))*10)
    c = np.cos(np.cumsum(f)) * amp
    s = np.sin(np.cumsum(f)) * amp
    return (c,s)
```
